```python
#importing and cleaning dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv("data.csv")


#key={'Malignant':2.0, 'Benign':'1.0'}

for i in df.index:
 if df.loc[i,"diagnosis"]=='M':
   df.loc[i,"diagnosis"]=float(2.0)
 else:
   df.loc[i,"diagnosis"]=float(1.0)

del df[df. columns[-1]]
del df[df.columns[0]]
#deleted first column containing patient IDs and last column containing 
#NA values

df
```

#LINEAR REGRESSION

```python
#predicting diagnosis with 19 features of the cell
X = df[['radius_mean','texture_mean','perimeter_mean', 'area_mean','smoothness_mean','compactness_mean','concavity_mean','concave points_mean','symmetry_mean','radius_worst','texture_worst','perimeter_worst','area_worst','smoothness_worst','compactness_worst','concavity_worst','concave points_worst','symmetry_worst','fractal_dimension_worst']]
y= df['diagnosis']
from sklearn import linear_model 
regr=linear_model.LinearRegression()
regr.fit(X.values,y)
#"x.values" needed to be specified cos just x is going to include the header 
print(regr.coef_)

predicteddiagnosis = regr.predict([[20.57,17.77,132.90,1326.0,0.08474,0.07864,0.08690,0.07017,0.1812,24.990,23.41,158.80,1956.0,0.12380,0.18660,0.2416,0.1860,0.2750,0.08902]])
print(predicteddiagnosis)
```

```python
#checking r2 score
from sklearn.metrics import r2_score

print(r2_score(y, regr.predict(X.values)))
```

#PREDICTING WITH POLYNOMIAL REGRESSION

```python
#Using polynomial regression model to predict relationships between concave points mean and perimeter mean
x=df["concave points_mean"]
y=df["perimeter_mean"]

trains_x = x[:455]
trains_y = y[:455]

tests_x = x[455:]
tests_y = y[455:]

mymodel = np.poly1d(np.polyfit(trains_x, trains_y, 4))


myline = np.linspace(0, 40, 455)

plt.scatter(trains_x, trains_y)
plt.plot(myline, mymodel(myline))
plt.show()
```

```python
#now checking if it fits testing set
r2 = r2_score(tests_y, mymodel(tests_x))

print(r2)
```
